```python
# compare two variables 
# cost vs rating 
import seaborn as sns 
```

```python
plt.figure()
sns.scatterplot(x="cost",y="rate",data=df,alpha=0.6)
plt.title("cost vs rate",fontsize=14,fontweight='bold')
plt.xlabel("Cost")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure()
plt.hist(df['rate'].dropna(),bins=20)
plt.title("distribution of restaurant ratings",fontsize=14,fontweight='bold')
plt.xlabel("Rating")
plt.ylabel("Number of restaurant")
plt.grid(True)
plt.show()
```

```python
# bar chart 
top_locations=df['location'].value_counts().head(10)
print(top_locations)
```

```python
plt.figure()
top_locations.plot(kind="bar")
plt.title("Top 10 Locations by Restaurant count",fontsize=14,fontweight='bold')
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# online order vs rating (box plot)
plt.figure()
sns.boxplot(
    x="online_order",
    y="rate",
    data=df
)
plt.title("online_order impact on rating",fontsize=14,fontweight='bold')
plt.xlabel=("online_oder")
plt.ylabel=("rating")
plt.show()
```

```python
# votes distribution 
plt.figure()
plt.hist(df["votes"],bins=30)
plt.title("votes distribution",fontsize=14,fontweight='bold')
plt.xlabel="votes"
plt.ylabel="count"
plt.show()
```

```python
# correlation heatmap()
plt.figure()
corr=df[['rate','votes','cost']].corr()
sns.heatmap(corr,annot=True,fmt=".2f")
plt.title("correlation heatmap",fontsize=14,fontweight='bold')
plt.show()
```
